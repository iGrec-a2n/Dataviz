# Exercie 4 - Visualisation Barplot

```python
import pandas as pd
import numpy as np

flipper=pd.read_csv('../../data/flipper_games.csv')
flipper
games_clean = flipper.drop_duplicates()
games_clean = games_clean[games_clean["game_duration_s"] > 0]
games_clean = games_clean[games_clean["tilt_risk_pct"].between(0, 100)]
games_clean = games_clean[games_clean["pause_s"] >= 0]
games_clean["is_error"] = np.where(
    games_clean["game_status"] == "failed",  # ou la valeur qui indique une erreur
    1,
    0
)
zone_kpi = games_clean.groupby("arcade_zone").agg(
    nb_parties=("arcade_zone", "count"),
    nb_unique_player=("player_id", "nunique"),
    error_rate=("is_error", "mean"),
    average_pause=("pause_s", "mean")
).sort_values("error_rate", ascending=False)

zone_kpi
```

## 1. Barplot de Taux d'erreur / zone_kpi

```python
import seaborn as sns
import matplotlib.pyplot as plt

zone_kpi_sorted = zone_kpi.sort_values("error_rate", ascending=False).reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=zone_kpi_sorted,
    x="arcade_zone",
    y="error_rate",
    color="steelblue"
)
ax.set_title("Taux d'erreur par zone d'arcade")
ax.set_xlabel("Zone d'arcade")
ax.set_ylabel("Taux d'erreur")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
**Interprétation métier**  
La zone C2 (et dans une moindre mesure C3) apparaît comme l'une des plus critiques avec un taux d'erreur parmi les plus élevés.  
Une action possible serait d'analyser plus finement cette zone (problèmes de machine, règles peu claires, disposition) et de tester des actions correctives ciblées : maintenance renforcée, amélioration de l'ergonomie ou clarification des consignes aux joueurs.
```
