# Exercie 2

`````python
import numpy as np
import pandas as pd

flippers_raw = pd.read_csv('../../data/flipper_games.csv')
flippers_raw.info()
`````

---
## Suppression des doublons

`````python
f_no_dup = flippers_raw.drop_duplicates()
f_no_dup
`````

---
## On conserve les parties avec ```game_duration_s>0```

`````python
f_game_duration = (f_no_dup[f_no_dup['game_duration_s']>0])
f_game_duration
`````

## On conserve les lignes avec ```0<tilt_risk_pct<100```

`````python
type(f_game_duration)
`````

`````python
f_tilt_risk=f_game_duration[f_game_duration["tilt_risk_pct"].between(0,100)]
f_tilt_risk
`````

---
## On conserve les ligne avec ```pause_s >= 0````

`````python
f_clean=f_tilt_risk[f_tilt_risk["pause_s"]>=0]
f_clean
`````

---
## Vérification du nombre de lignes

`````python
before=flippers_raw.shape[0]
after=f_clean.shape[0]
print("before: ", before,"lignes")
print("After: ", after, "lignes")
print("Nombre de lignes retirées: ", (before - after))
`````
