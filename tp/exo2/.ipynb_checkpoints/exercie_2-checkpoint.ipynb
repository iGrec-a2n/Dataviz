````python
import numpy as np
import pandas as pd

flippers_raw = pd.read_csv('../../data/flipper_games.csv')
flippers_raw.info()
````

---
## Suppression des doublons

````python
flippers = flippers_raw.drop_duplicates()
flippers.info()
````

---
## On conserve les parties avec ```game_duration_s>0```

````python
flippers = (flippers[flippers['game_duration_s']>0])
flippers
````

## On conserve les lignes avec ```0<tilt_risk_pct<100```

````python
type(flippers)
````

````python
flippers=flippers[flippers["tilt_risk_pct"].between(0,100)]
flippers
````
